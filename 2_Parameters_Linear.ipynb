# Effect of different parameters

```python
import numpy as np
import matplotlib.pyplot as plt

from glacier import *

dt = 0.1
t = 500
calving = True

save = False
loc = "plots/parameter-test/"
```

## Alpha

```python
a = LinearBedModel(calving=calving)
b = LinearBedModel(calving=calving)
c = LinearBedModel(calving=calving)
d = LinearBedModel(calving=calving)
e = LinearBedModel(calving=calving)
a.alpha = 3
b.alpha = 2
c.alpha = 4
d.alpha = 1
e.alpha = 5

glaciers = [a,b,c,d,e]

for glacier in glaciers:
    print(glacier)
    glacier.E = 800
    glacier.integrate(dt, t)
    glacier.E = 1000
    glacier.integrate(dt, t)
    glacier.E = 800
    glacier.integrate(dt, t)
    if save: 
        glacier.plot(E=True,C=True,show=False,filename=loc+"alpha"+str(glacier.alpha))
    plt.clf()
    plt.close()
    
plt.figure(figsize=(12,4), dpi=200)   

plt.plot(a.t, a.L, label="alpha = {}".format(a.alpha))
plt.plot(b.t, b.L, label="alpha = {}".format(b.alpha))
plt.plot(c.t, c.L, label="alpha = {}".format(c.alpha))
plt.plot(d.t, d.L, label="alpha = {}".format(d.alpha))
plt.plot(e.t, e.L, label="alpha = {}".format(e.alpha))

plt.grid()
plt.legend(bbox_to_anchor=(1, 1), loc="upper left")
plt.xlabel("Time [year]")
plt.ylabel("Length [meter]")
plt.tight_layout()
if calving: plt.ylim([42000, 47000])
elif not calving: plt.ylim([60000, 75000])
plt.show()
```
